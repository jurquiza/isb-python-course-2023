```python
plt.figure(dpi=600)
for power in range(0,5,1):
    x = sp.linspace(0,10,10)
    y = x**power+1
    #o is the marker type
    #here we are plotting x vs y after calculating the power function
    #subplo will create individual plots for each iteration in the for loop
    plt.subplot(2,3,power+1)
    #we can change the colour of the "data" by providing a rule in this case dividin one by the power 
    plt.plot(x,y, "o" ,color=(0.1,1./(power+1),0.5))
    plt.yscale("log")
    plt.ylim(1,1e3)
    plt.tight_layout()
#plt.plot(x,y, "o" ,color=(0.1,1./(power+1),0.5))
```
